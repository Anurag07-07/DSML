```python
# closures

def main_welcome(func):
  msg = "Welcome"
  def sub_welcome_method():
    print("IDK")
    func()
    print("IDK2")
  return sub_welcome_method()
```

```python
main_welcome(print)
```

```python
def my_decorator(func):
  def wrapper():
    print("dasd")
    func()
    print("dasd")

  return wrapper()
```

```python
@my_decorator
def say_hello():
  print("Hello")
```

```python
def repeat(n):
  def decorator(func):
    def wrapper(*args,**kwargs):
      for _ in range(n):
        func(*args,**kwargs)
    return wrapper
  return decorator
```

```python
@repeat(3)
def say_hello():
  print("Hello")  

```

```python
say_hello()
```
