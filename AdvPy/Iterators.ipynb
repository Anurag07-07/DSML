# Iterators 

```python
my_list = [1,2,3,4,5,6]

# Iterator
iterator = iter(my_list)
print(type(iterator))
```

```python
## Iterator thorugh all the element
next(iterator)
```

```python
try:
  next(iterator)
except StopIteration:
  print("There is no elements in the iterator")
```

# Generator 

```python
def square(n):
  for i in range(3):
    yield i**2
```

```python
square(3)
```

```python
for i in square(3):
  print(i)
```

```python
a = square(3)
next(a)
```

```python
def my_generator():
  yield 1
  yield 2
  yield 3
```

```python
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
  print(val)
```

# Reading Large File

```python
def read_large_file(file_path):
  with open(file_path,'r') as file:
    for line in file:
      yield line
```

```python
file_path = 'a.txt'
for line in read_large_file(file_path):
  print(line.strip())
```
