```python
# Exception Handling 
try:
  a = b
except NameError as ex:
  print("Variable has not been assigned")
  print(ex)
```

```python
try:
  1/0
except ZeroDivisionError as ex:
  print(ex)
  print("Please enter any other number in deno")
```

```python
try:
  result = 1/2
  a = b
except ZeroDivisionError as ex:
  print(ex)
  print("Please enter any other number in deno")
except Exception as ex1:
  print(ex1)
```

```python
try:
  num = int(input("Enter a number"))
  result = 10/num
except ValueError:
  print("This is not a valid Number")
except ZeroDivisionError:
  print("Enter Deno not Equal zero")
except Exception as ex:
  print(ex)
```

```python
try:
  num = int(input("Enter a number"))
  result = 10/num
except ValueError:
  print("This is not a valid Number")
except ZeroDivisionError:
  print("Enter Deno not Equal zero")
except Exception as ex:
  print(ex)
else:
  print(f" the result is {result}")
```

```python
try:
  num = int(input("Enter a number"))
  result = 10/num
except ValueError:
  print("This is not a valid Number")
except ZeroDivisionError:
  print("Enter Deno not Equal zero")
except Exception as ex:
  print(ex)
else:
  print(f" the result is {result}")
finally:
  print("Execution Completed")
```

```python
try:
  file = open('example1.txt','r')
  content = file.read()
  print(content)

except FileNotFoundError:
  print("the file is not there")
finally:
  if 'file' in locals() and not file.closed():
    file.close()
    print("File Close")
```
