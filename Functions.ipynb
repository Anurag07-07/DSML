```python
def convert_temp(temp,unit):
    """This Function Convert Temp bw c to f"""
    if unit=='C':
        return temp * 9/5 + 32 ##Celcius to F
    elif unit=='F':
        return (temp-32)*(5/9)
    else:
        return None

convert_temp(56,'C')
convert_temp(12,'F')
convert_temp(12,'D')
```

```python
def is_strong(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*/*-+' for char in password):
        return False
    return True

print(is_strong('772002'))
```

```python
def calculate_total_cost(cart):
    total_cost = 0
    for  item in cart:
        total_cost+=item['price']*item['quantity']
        
    return total_cost

cart=[
    {'name':"Apple","price":0.5,"quantity":4},
    {'name':"Orange","price":0.3,"quantity":2},
    {'name':"Peach","price":0.8,"quantity":5},
]

ans = calculate_total_cost(cart)
print(ans)
```

```python
def ispalind(s):
    s = s.lower().replace(" ","")
    return s==s[::-1]

ans = ispalind("A man a plan a canal Panama")
print(ans)
```

```python
def fac(n):
    if n==0:
        return 1
    return n*fac(n-1)

print(fac(5))
```

```python
def count_word(file_path):
    word_count = {}
    with open(file_path,'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word =word.lower().strip()
                word_count[word] = word_count.get(word,0)+1

    return word_count

file_path = "a.txt"
ans = count_word(file_path)
print(ans)
```

# Filter Function

```python
lst = [1,2,3,1,5,6,4,5,6]
even = lambda a:a%2==0
print(list(filter(even,lst)))
```

```python
people = [
 {"name":"a","age":18},
 {"name":"b","age":20},
 {"name":"c","age":12},
]

def filter_age(people):
    return people['age']>17


ans = list(filter(filter_age,people))
print(ans)
```
