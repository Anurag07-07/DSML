# Function is Block of Code which runs when it is called
# In Python a function is defined with the def keyword
# Functions can accept any number of arguments of any type
# Functions can optionally return data as well.This data may also be of any type.You can return more than one value as well
# Arguments can have default Values as well
# While Calling Functions The user may define arguments as keyword pairs;

```python
def greet(name="You Know you Know"):
    print(f"Hello {name}")

greet("Anurag")
greet()
```

# Cube of Number

```python
def cube(x):
    return x**3

print(cube(3))
```

```python
# Multiple Return
def bio():
    return "Anurag",21,False

name,age,Relation = bio()
print("Name",name)
print("Age",age)
print("Status",Relation)
```

```python
def data(name,skills):
    print(f" My name is {name} and skills {skills}")

data(skills=["Javascript","Typescript"],name="Anurag Raj")
```

```python
m =  10
def square(m): # Local variable Shadowed the Global Variable
    return m**2

def multipleoffive(): # Use Global Variable
    global m
    print(m**5)

print(square(5))
multipleoffive()
```

# Map Function

```python
def square(x):
    return x**2
a = [1,2,3,4,5]

b = list(map(square,a)) 
print(b)
```

# Lambda Function

```python
square = lambda x:x**2
print(square(50))
```

```python
cube = lambda x:x**3
print(cube(50))
```

```python
multiply = lambda a,b:a*b
print(multiply(50,49))
```

```python
d = lambda b:[b**2,b+b]
print(d(10))
```
