# List Comprehension

```python
res = [ i**2 for i in range(1,101)]
for i in res:
    print(i,end=" ")    
```

```python
# Filter out multiple of 5
[i for i in res if i%5==0]
```

# Tuples
# Tuples are used to store multiple items in a single variable.

# A tuple is a collection which is ordered and immutable (unchangeable).

# Tuples are written with round brackets.

# Tuple items are indexed; the first item has index [0], the second item has index [1], etc.

# Tuples can be created with the tuple constructor: tuple().

# Tuples can also be created by simply writing multiple values with comma separation.

# Tuples are processed faster than lists in Python.

```python
vowels = ("a","e","i","o","u")
print(vowels)
```

```python
print(type(vowels))
```

```python
planets=("Mercury","Venus","Hmar Gola","Mars","Jupiter","Saturn")
print(planets)
```

```python
#decleration
a = tuple()
```

```python
a = tuple([1,2,4,5,6,3])
print(a)
```

# Value change in tuple
# We can't directly change value in tuple so we convert it into list
# Tuple is faster than list

```python
a = (1,2,3,4)
b = list(a)
b[0] = 10
a = tuple(b)
print(a)
```

```python
a = [1,2,3]
b = ('a','b',a)
a.append([1,2,3])
print(b)
```

# Sorted 

```python
a = (1,4,10,12,14,18,20)

flag = True

for i in range(len(a)-1):
    if a[i]>a[i+1]:
        flag=False

if flag:
    print("Sorted")
else:
    print("Not Sorted")
        
```

# Function Args 

```python
def product(*args):
    print(args,type(args))
    p = 1
    for i in args:
        p = p*i

    print(p)

product(1,2,3,4,5,6)
```
