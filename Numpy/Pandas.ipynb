# Pandas

```python
import pandas as pd
```

```python
## Pandas Series one d array that holds any type of data
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
# Create a Series from Dict
data = {'a':1,'b':2,'c':3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index = ['a','b','c']

pd.Series(data,index)
```

```python
# Dataframe
## Create a Dataframe from a dictionary of list
data = {
  'Name':["AJ","BJ","CJ","DJ"],
  "Age":[25,30,45,102],
  "City":["IDK1","IDK2","IDK3","IDK4"]
}

df = pd.DataFrame(data)
print(type(df))
print(df)
```

```python
# Create a Data Frame a Lsit of Dict
data = [
  {"Name":"AJ","age":25,"city":"NY"},
  {"Name":"BJ","age":22,"city":"Dallas"},
  {"Name":"CJ","age":26,"city":"Texas"},
  {"Name":"DJ","age":28,"city":"Calfornia"},
]

df = pd.DataFrame(data)
print(df)
print(type(df))
```
