## Classes are the user-defined blueprints that help us create an object

```python
class Car:
    def describe(self):
        print("Car name :",self.name)
        print("Car name :",self.color)
        print("Car name :",self.age)

    def isBanned(self):
        if self.age > 15:
            return True
        else:
            return False
```

```python
benz = Car()
benz.name = "Mercedes Benz"
benz.color = "Silver"
benz.age = 2
benz.describe()
```

# Constructors

```python
class Car:
    def __init__(self,name,color,company):
        self.name = name
        self.color = color
        self.company = company

    def describe(self):
        print("Car name :",self.name)
        print("Car name :",self.color)
        print("Car name :",self.company)
```

```python
mac = Car("P1","Orange","Maclaren")
mac.describe()
```

# Class inside Class

```python
class Car:
    class Seat:
        def __init__(self,color="grey"):
            self.color = color

    def __init__(self):
        self.name  = "Anurag"

    def drive(self):
        s = self.Seat("Red")
        print(s)
```

```python
c1 = Car()
c1.drive()
```
