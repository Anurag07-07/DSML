# Encapsulation

```python
class Person:
  def __init__(self,name,age):
    self.name = name
    self.age = age



person = Person("someone",96)
print(person.name)
```

```python
class Person:
  def __init__(self,name,age,gender):
    self.__name = name #private variables
    self.__age = age  
    self.gender = gender 


person = Person("someone",96,"IDK")
dir(person)
```

```python
class Person:
  def __init__(self,name,age,gender):
    self._name = name #protected variables
    self._age = age  
    self.gender = gender 

class Emp(Person):
  def __init__(self, name, age, gender):
    super().__init__(name, age, gender)


emp = Emp("Someone",34,"IDK")
print(emp._name)
print(emp.gender)
print(emp._age)
```

```python
# En with getter and setter
class Person:
  def __init__(self,name,age):
    self.__name = name  # Private access modifier or variable
    self.__age = age ## Private variable

  def get_data(self):
    return self.__name
  
  ## Setter method for name
  def set_name(self,name):
    self.__name = name

  def set_age(self,age):
    if age>0:
      self.__age = age
    else:
      print("Enter the valid data")

  def get_age(self):
    return self.__age


person = Person("someone",89)
print(person.get_data())
print(person.get_age())
```
