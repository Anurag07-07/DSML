# Inheritance

```python
class Car:
  def __init__(self,windows,doors,engine):
    self.windows = windows
    self.doors = doors
    self.engine = engine


  def drive(self):
    print(f"The Person is Drive the {self.engine} Car")
```

```python
maclaren = Car(2,2,"V8")
maclaren.drive()
```

```python
class Koiensegg(Car):
  def __init__(self, windows, doors, engine,speed):
    super().__init__(windows, doors, engine)
    self.speed = speed

  def speed_testing(self):
    print(f"The Top speed is {self.speed} KM/H")
```

```python
Regera = Koiensegg(2,2,"V12",400)
Regera.speed_testing()
```

# Multiple Inheritance 

```python
class Animal:
  def __init__(self,name):
    self.name = name

  def speak(self):
    print("Hello1")

class Pet1:
  def __init__(self,owner):
    self.owner = owner

  def speak(self):
    print("Owner is the pet of pet")


class Dog(Animal,Pet1):
  def __init__(self, name,owner):
    Animal.__init__(self,name)
    Pet1.__init__(self,owner)

  def speak(self):
    return f"{self.name} say IDK"



dogy = Dog("Bruno","Buddy")
dogy.speak()
```
