# Method Overiding

```python
class Animal:
  def speak(self):
    return "Sound of the Animal"
  
class Dog(Animal):
  def speak(self):
    return "Woof!"
  
def animal_speak(animal):
  print(animal.speak())

dog = Dog()
animal = Animal()

animal_speak(dog)
```

```python
# Polymorphism with Fucntions and Methods
# base class
class Shape:
  def area(self):
    return "The area of the figure"
  
#Derived class 1
class Rectangle(Shape):
  def __init__(self,width,height):
    self.width = width
    self.height = height

  def area(self):
    return self.width*self.height
  
# Derived class 2
class Circle(Shape):
  def __init__(self,radius):
    self.radius = radius

  def area(self):
    return 3.14*(self.radius)**2
  


def print_area(shape):
   print(f"The area is {shape.area()}")


circ = Circle(4)
rect = Rectangle(4,5)

print_area(circ)
print_area(rect)
```

```python
# Abstract class
from abc import ABC,abstractmethod

# Define abstract class

class Vehicle(ABC):
  @abstractmethod
  def start_engine(self):
    pass


class Car(Vehicle):
  def start_engine(self):
    return "Car engine Started"
  
class MotorCycle(Vehicle):
  def start_engine(self):
    return "Motor engine started"
  


def main(value):
  print(value.start_engine())

car = Car()
motor = MotorCycle()

main(car)
main(motor)
```
