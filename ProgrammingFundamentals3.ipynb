# Check Prime 

```python
n = int(input("Enter The Number"))

flag = True

for i in range(2,n):
    if n%i==0:
        flag = False
        break


if flag:
    print("Prime Number")
else: 
    print("Not Prime")
```

# Linear Search

```python
a = [2,6,9,3,4,1,5]

print(a.index(9)) # Linear Search Inbuilt

flag = False
target = 9
for i in range(0,len(a)):
    if a[i]==target:
        flag = True
        break;

if flag:
    print("Present")
else:
    print("Not Present")
```

```python
nums = [4,5,8,1,9,2]
start = 0
end = len(nums)-1
while start<end:
    nums[start],nums[end] = nums[end],nums[start]
    start+=1
    end-=1

print(nums)
```
