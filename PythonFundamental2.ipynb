# Print Sum of Digits 

```python
n = int(input("Enter the Number"))
sum = 0
while n>0:
    last=n%10
    sum = sum + last
    n = n//10
print(sum)
```

```python
'''
0001000
0023200
0345430
4567654
'''
n = int(input("Enter The Row"))

for i in range(1,n+1):
    # print 0 row-1 times first 
    print("0"*(n-i),end="")
    # start form i goes 2i-1
    for j in range(i,2*i):
        print(j,end="")
    for j in range(2*i-2,i-1,-1):
        print(j,end="")
    #print zero again
    print("0"*(n-i))
```
