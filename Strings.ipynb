# Function kwargs

## keyword arguments

### So basically kwargs store data in form of key valuepair

```python
def printfunc(**kwargs):
    print(kwargs)

printfunc(name="Anurag",name1="Something")    
```

```python
def hogwarts(house1,house2,house3,house4,*args,**kwargs):
    print("Houses are ",house1,house2,house3,house4)
    print()
    print("Teachers:",args)
    print()
    print("Some popular students:",kwargs)

hogwarts('a','b','c','d','e','f','g',h='h',i='i')    
```

# Format Strings 

```python
apple_count = 4
mango_count = 10
orange_count = 6

# generate a string that says, "I have 4 apples,10 oranges and 5 guavas"

s = f"I have {apple_count} apples,{mango_count} mangos, and {orange_count} Oranges"
print(s)
```

```python
s = "I have {} apples,{} mangos, and {} Oranges".format(apple_count,mango_count,orange_count)
s1 = "I have {2} apples,{1} mangos, and {0} Oranges".format(apple_count,mango_count,orange_count) # we can also give the string
print(s)
print(s1)
```
