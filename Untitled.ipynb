# File Operations 

```python
# Read a whole file

with open('a.txt','r') as file:
    content = file.read()
    print(content)
```

```python
# Read a file line by line
with open('a.txt','r') as file:
    for line in file:
        print(line.strip()) # .strip() remove new line character
```

```python
# Writing the file (Overwritting)

with open('a.txt','w') as file:
    file.write('Hello World\n')
    file.write('this is a new line\n')
```

```python
# Write file without overwriting
with open('a.txt','a') as file: 
    file.write("Append operation taking place")
```

```python
# Writing list of lines to a file
lines = ['First line\n',"Second line\n"]
with open('a.txt','a') as file:
    file.writelines(lines)
```

```python
data = b'\x00\x01\x02\x03'
with open('a.bin','wb') as file:
    file.write(data)
```

```python
with open('a.bin','rb') as file:
    a = file.read()
    print(a)
```

```python
# bfile is created if not there. use for reading and writing
with open ('a.txt','w+') as file:
    file.write("Hello world\n")
    file.write("Hello world\n")
    ## Move the file cursor to the beginning
    file.seek(0)
    ## Read the content of the file
    content = file.read()
    print(content)
```

```python
import csv

with open('example.csv',mode='w',newline="") as file:
  writer = csv.writer(file)
  writer.writerow(['name','age'])
  writer.writerow(['Krish',32])

with open('example.csv',mode='r') as file:
  reader = csv.reader(file)
  for row in reader:
    print(row)
```
